```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df= pd.read_csv('cost_revenue_dirty.csv')
df.head()
```

```python
df.isna().values.any()
```

```python
duplicated_rows=df.duplicated().values.any()
```

```python
df.shape
```

```python
duplicated_rows = df[df.duplicated()]
len(duplicated_rows)
```

```python
df.info()
```

```python
chars_to_remove = [',', '$']
columns_to_clean = ['USD_Production_Budget', 
                    'USD_Worldwide_Gross',
                    'USD_Domestic_Gross']
for col in columns_to_clean:
  for char in chars_to_remove:
    # Replace each character with an empty string
    df[col]=df[col].astype(str).str.replace(char,"")
    # Convert column to a numeric data type
  df[col] = pd.to_numeric(df[col])
```

```python
df.head()
```

```python
df.Release_Date = pd.to_datetime(df.Release_Date)
```

```python
scrape_date= pd.Timestamp('2018-5-1')
future_releases=df[df.Release_Date>=scrape_date]
len(future_releases)
```

```python
data_clean = df.drop(future_releases.index)
```
